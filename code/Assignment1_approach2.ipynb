```python
%%capture
import torch
from diffusers import DiffusionPipeline
from diffusers.models.attention_processor import AttnProcessor2_0
from matplotlib import pyplot as plt

pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16").to("cuda")
```

```python

object_ = "Perfume" #@param ["Shoe", "Sneaker", "Bottle", "Cup", "Sandal", "Perfume", "Toy", "Sunglasses", "Car", "Water Bottle", "Chair", "Office Chair", "Can", "Cap", "Hat", "Couch", "Wristwatch", "Glass", "Bag", "Handbag", "Baggage", "Suitcase", "Headphones", "Jar", "Vase"]
lighting_condition = "natural" #@param {type:"string"}
background = "nature" #@param {type:"string"}
position = "front" #@param ["front","front_left","front_right","right","left","bottom","top","rear","rear_left","rear_right"]
```

```python
if background == "None":
  prompt = f"realistic photoshoot visual of singlular {object_} from a {position} position in {lighting_condition} ligthing condition"
else:
  prompt = f"realistic photoshoot visual of singlular {object_} from a {position} position in {lighting_condition} ligthing condition with a {background} background of the image"
image = pipe(prompt,num_inference_steps=4, guidance_scale=0.0).images[0]
plt.axis("off")
plt.imshow(image)
```
